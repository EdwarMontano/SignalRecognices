DENSIDAD DE PROBABILIDAD INCONDICIONAL p(X)

Genera la probabilidad de iobservar el valor de x en una medici√≥n respecto a todo el conjunto de datos, sin importar la pertencia a las clases

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

from sklearn.neighbors import KernelDensity

# ----------------------------------------------------------------------

np.random.seed(1)
# Genera Datos Aleatorios Clase 1 m=3, sigm= 1.5
N1 = 1000
m1=3
sigm1=1.5
X1=np.random.normal(m1, sigm1, N1)

# Genera Datos Aleatorios Clase 2: m=7, sigm= 2.5
N2 = 1000
m2=7
sigm2=2.5
X2=np.random.normal(m2, sigm2, N2)

# Graficar puntos Class 1
fig, axs = plt.subplots(2,1)
axs=fig.add_subplot(2,1,1)
axs.scatter(X1,np.zeros(X1.size) ,marker = 'o', c='blue')
#plt.title("Data set - Class 1")
plt.xlim(-2,18)
axs.legend(['Class 1: '+ 'm='+ str(m1)+ '  sig='+ str(sigm1) ])
axs.set_xlabel('X values')
axs.set_ylabel('Class 1')

# Graficar puntos Class 2
axs=fig.add_subplot(2,1,2)
axs.scatter(X2,np.zeros(X2.size) ,marker = 'o', c='red')
#plt.title("Data set - Class 2")
plt.xlim(-2,18)
axs.legend(['Class 2: '+ 'm='+str(m2)+ '  sig='+ str(sigm2) ])
axs.set_xlabel('X values')
axs.set_ylabel('Class 2')
   
```

Generar un solo data set. Luego se obtiene p(x) del histograma 

```python
# Obtener un solo data set

X=np.concatenate((X1, X2))

min_X=np.min(X)
max_X=np.max(X)

Nbin=int((N1+N2)/10)

delta=(max_X-min_X)/Nbin


px= plt.hist(X, bins=Nbin ,density=True,color='g', histtype='step' ) 
plt.title("Incond. Probability p(x) ")
plt.xlabel('X values')
plt.ylabel('p(x)')

# Plot a normal distribution:  m1  sigma1

min_X1=np.min(X1)
max_X1=np.max(X1)
Nbin1=int((N1)/10)
delta1=(max_X1-min_X1)/Nbin1

ejeX1=np.arange(min_X1, max_X1,delta1)
plt.plot(ejeX1,norm.pdf(ejeX1,m1,sigm1), color='blue')

# Plot a normal distribution: :  m2  sigma2
min_X2=np.min(X2)
max_X2=np.max(X2)
Nbin2=int((N2)/10)
delta2=(max_X2-min_X2)/Nbin2
ejeX2=np.arange(min_X2, max_X2,delta2 )
plt.plot(ejeX2,norm.pdf(ejeX2,m2,sigm2),color='red')

plt.show()
```
